```python
import json
import numpy as np
import pandas as pd
```

```python
with open("train_no_dup.json") as f:
    train_no_dup = json.load(f)
train_no_dup
```

```python
for outfit in train_no_dup:
    products = []
    for item in outfit['items']:
        products.append(int(item['item_id']))
    outfit['products'] = products
```

```python
train = pd.DataFrame(train_no_dup)
train.head()
```

```python
train.rename(columns = {'set_id':'outfit_id', 'items':'items', 'products':'products'}, inplace = True)
train = train[['products', 'outfit_id']]
train.head()
```

```python
path = "../../../../../farfetch_challenge_utils2/data"

train.to_parquet(path + "/manual_outfits_train_baseline.parquet")
```
