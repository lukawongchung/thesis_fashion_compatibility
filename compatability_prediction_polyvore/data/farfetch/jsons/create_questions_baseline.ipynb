```python
import json
import numpy as np
import pandas as pd
```

```python
with open("fill_in_blank_test.json") as f:
    fill_in_blank_test = json.load(f)
fill_in_blank_test
```

```python
full_outfits = pd.read_parquet("manual_outfits.parquet")
full_outfits.iloc[36194]['products']
```

```python
full_outfits = pd.read_parquet("manual_outfits.parquet")
for question in fill_in_blank_test:
    outfit = int(question['question'][0].split('_')[0])
    question['outfit_id'] = outfit
    question['question'] = full_outfits.iloc[outfit]['products']
    missing_product = full_outfits.iloc[outfit]['products'][question['blank_position'] - 1]
    question['question'].pop(question['blank_position'] - 1)
    question['blank_position'] = missing_product
```

```python
fill_in_blank_test
```

```python
full_outfits = pd.read_parquet("manual_outfits.parquet")
for question in fill_in_blank_test:
    for i in range(len(question['answers'])):
        answer = question['answers'][i]
        outfit, index = int(answer.split('_')[0]), int(answer.split('_')[1])
        items = full_outfits.iloc[outfit]['products']
        question['answers'][i] = items[index - 1]
```
