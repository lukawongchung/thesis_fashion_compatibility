```python
import json
import numpy as np
import pandas as pd
```

```python
with open("test_no_dup.json") as f:
    test_no_dup = json.load(f)
```

```python
test_df = pd.DataFrame(test_no_dup)
test_df.head()
```

```python
def return_item_ids(lst_of_dct):
    return [str(item["item_id"]) + '_' + str(item["index"]) for item in lst_of_dct]
```

```python
test_df["item_ids"] = test_df.apply(lambda row : return_item_ids(row["items"]), axis=1)
```

```python
def remove_item(item_ids):
    remove = random.choice(item_ids)
    item_ids.remove(remove)
    return item_ids
```

```python
import random

test_df['question'] = test_df['item_ids']
test_df.apply(lambda row : remove_item(row['question']), axis=1)
test_df["item_ids"] = test_df.apply(lambda row : return_item_ids(row["items"]), axis=1)
test_df['missing_product'] = test_df.apply(lambda row : list(set(row['item_ids']) - (set(row['question'])))[0], axis=1)
test_df.head()
```

```python
def all_outfits(lst_of_dct):
    return [item["item_id"] for item in lst_of_dct]
```

```python
# Get all items that are in the test_set
all_outfits = test_df.apply(lambda row : all_outfits(row['items']), axis=1)
all_outfits.head()
```

```python
import itertools

all_outfits_list = test_df['item_ids'].tolist()
all_items_list = list(itertools.chain.from_iterable(all_outfits_list))
all_items_list = list(dict.fromkeys(all_items_list))
all_items_series = pd.Series(all_items_list)
all_items_series.head()
```

```python
def candidates(row):
    n = 3
    c = all_items_series.sample(n).unique().tolist()
    c.append(row["missing_product"])
    return list(set(c))

test_df["answers"] = test_df.apply(lambda row: candidates(row), axis=1)
test_df.head()
```

```python
test_df['blank_position'] = test_df.apply(lambda row : int(row['missing_product'][-1]), axis=1)
test_df.head()
```

```python
fitb_df = test_df.drop(['set_id', 'items', 'item_ids', 'missing_product'], axis=1)
fitb_df.head()
```

```python
fill_in_blank_test = fitb_df.to_dict("records")
fill_in_blank_test[0]
```

```python
with open('fill_in_blank_test.json', 'w') as f:
    json.dump(fill_in_blank_test, f)
```
