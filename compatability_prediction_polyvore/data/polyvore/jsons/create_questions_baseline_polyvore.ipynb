```python
import json
import numpy as np
import pandas as pd
```

```python
with open("test_no_dup.json") as f:
    test_no_dup = json.load(f)
test_no_dup
```

```python
map_id2their = {}

for outfit in test_no_dup:
    for item in outfit['items']:
        # get id from the image url
        _, id = item['image'].split('id=')
        id = int(id)
        #cat_dict[id] = item['categoryid']
        map_id2their['{}_{}'.format(outfit['set_id'], item['index'])] = id
map_id2their
```

```python
with open("fill_in_blank_test.json") as f:
    fill_in_blank_test = json.load(f)
fill_in_blank_test
```

```python
for question in fill_in_blank_test:
    question['outfit_id'] = int(question['question'][0].split('_')[0])
    question['question'] = [map_id2their[item] for item in question['question']]
    question['answers'] = [map_id2their[item] for item in question['answers']]
    question['blank_position'] = question['answers'][0]
```

```python
fill_in_blank_test
```

```python
import random 
for question in fill_in_blank_test:
    random.shuffle(question['answers'])
```

```python
fill_in_blank_test
```

```python
test = pd.DataFrame(fill_in_blank_test)
test.head()
```

```python
test.rename(columns = {'outfit_id':'outfit_id', 'question':'incomplete_outfit', 'answers':'candidates', 'blank_position':'missing_product'}, inplace = True)
test.head()
```

```python
test_input = test[["outfit_id", "incomplete_outfit", "candidates"]]
test_input.head()
```

```python
test_output = test[["outfit_id", "missing_product"]]
test_output.head()
```

```python
path = "../../../../../farfetch_challenge_utils2/data"

test_input.to_parquet(path + "/manual_outfits_testinput_baseline_polyvore.parquet")
test_output.to_csv(path + "/manual_outfits_testoutput_baseline_polyvore.csv", index=False)
```
